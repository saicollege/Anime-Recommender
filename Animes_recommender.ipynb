```python
anime.loc[(anime["genre"]=="Hentai")& (anime["episodes"]=="Unknown"),"episodes"]="1"
anime.loc[(anime["type"]=="OVA")& (anime["episodes"]=="Unknown"),"episodes"]="1"
anime.loc[(anime["genre"]=="Movie")& (anime["episodes"]=="Unknown"),"episodes"]="1"
```

```python
sample_dataset={"Doraemon(1979)":1787, "One Piece":784,"Detective Conan":854,
                "Crayon Shin chan":942, "Fairy Tail":175,"Uchuu Kyoudai":99,
                "Code Geass: Hangyaku no Lelouch":25,"Shigatsu wa Kimi no Uso":22,"Naruto Shippuuden":500}
```

```python
for i,j in sample_dataset.items():
    anime.loc[anime["name"]==i,"episodes"] = j
```

```python
anime["episodes"]=anime["episodes"].map(lambda x:np.nan if x=="Unknown" else x)
```

```python
anime["episodes"].fillna(anime["episodes"].mode(),inplace=True)
```

```python
anime["rating"]=anime["rating"].astype(float)
```

```python
anime["rating"].fillna(anime["rating"].median(),inplace=True)
```

```python
anime
```

```python
pd.get_dummies(anime[["type"]]).head()
```

```python
anime["members"]=anime["members"].astype(float)
```

```python
anime_features=pd.concat([anime["genre"].str.get_dummies(sep=","),pd.get_dummies(anime[["type"]]),anime[["members"]],anime["episodes"]],axis=1)
```

```python
anime["name"]=anime["name"].map(lambda name:re.sub('[^A-Za-z0-9]+'," ",name))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.neighbors import NearestNeighbors


imputer = SimpleImputer(strategy='mean')
anime_features_imputed = imputer.fit_transform(anime_features)

nbrs = NearestNeighbors(n_neighbors=20, algorithm="ball_tree").fit(anime_features_imputed)
```

```python
distances, indices=nbrs.kneighbors(anime_features_imputed)
```

```python
def get_id_from_partial_name(partial):
    for index,row in all_anime.iterrows():
        if partial.lower() in row['name'].lower():
            print(row['name'],index)
```

```python
def get_index_from_name(name):
    return anime[anime["name"]==name].index.tolist()[0]
def print_similar_animes(query=None, id=None):
    if id is not None:
        found_id=id
    elif query is not None:
        found_id=get_index_from_name(query)
    else:
        print("Please provide either 'query' or 'id'.")
        return
    
    for similar_id in indices[found_id][1:]:
        print(anime.iloc[similar_id]["name"],"||",anime.iloc[similar_id]["genre"],"||",anime.iloc[similar_id]["type"])
print_similar_animes("Naruto")
```
